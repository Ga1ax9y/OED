Разбиваем выборку на интервалы

```python
def calculate_intervals(data):
    n = len(data)
    s = 1 + int(math.log2(n))
    min_val = min(data)
    max_val = max(data)
    interval_width = (max_val - min_val) / s
    intervals = [(min_val + i * interval_width, min_val + (i + 1) * interval_width) for i in range(s)]
    return intervals

intervals = calculate_intervals(filtered_values)
print("\nИнтервалы:")
for i, interval in enumerate(intervals):
    print(f"Интервал {i + 1}: [{interval[0]:.2f}, {interval[1]:.2f})")
```

```python
def plot_histogram_and_polygon(data, intervals):
    # Гистограмма
    plt.figure(figsize=(12, 6))
    plt.hist(data, bins=[interval[0] for interval in intervals] + [intervals[-1][1]], edgecolor="black", alpha=0.7, label="Гистограмма")

    # Полигон
    midpoints = [(interval[0] + interval[1]) / 2 for interval in intervals]
    frequencies = [sum(1 for x in data if interval[0] <= x <= interval[1]) for interval in intervals]
    plt.plot(midpoints, frequencies, marker="o", linestyle="-", color="red", label="Полигон")

    plt.title("Гистограмма и полигон распределения роста")
    plt.xlabel("Рост (см)")
    plt.ylabel("Частота")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_histogram_and_polygon(filtered_values, intervals)
```

Разбиваем выборку на s-1 и s+1 интервалов

```python

min_height = min(height_values)
max_height = max(height_values)


s = 8

def analyze_grouping(num_intervals, height_data, min_height, max_height):
    # Создаем интервалы равной ширины
    interval_width = (max_height - min_height) / num_intervals

    # Создаем границы интервалов
    interval_boundaries = [(min_height + i * interval_width, min_height + (i + 1) * interval_width)
                           for i in range(num_intervals)]

    # Распределяем страны по группам
    groups = {}
    for i, (lower, upper) in enumerate(interval_boundaries):
        group_name = f"Группа {i+1}: {lower:.1f}-{upper:.1f} см"
        groups[group_name] = []

        for country, height in height_data.items():
            # Для последнего интервала включаем верхнюю границу
            if i == num_intervals - 1:
                if lower <= height <= upper:
                    groups[group_name].append(height)
            else:
                if lower <= height < upper:
                    groups[group_name].append(height)

    # Рассчитываем общую среднюю и общую дисперсию
    height_values = list(height_data.values())
    overall_mean = np.mean(height_values)
    overall_variance = np.var(height_values, ddof=1)

    # Рассчитываем групповые дисперсии
    group_means = {}
    group_variances = {}
    group_sizes = {}

    for group_name, heights in groups.items():
        if heights:  # Проверяем, что группа не пустая
            group_means[group_name] = np.mean(heights)
            group_variances[group_name] = np.var(heights, ddof=1) if len(heights) > 1 else 0
            group_sizes[group_name] = len(heights)

    # Рассчитываем внутригрупповую дисперсию
    total_elements = sum(group_sizes.values())
    within_group_variance = sum(group_variances[g] * group_sizes[g] for g in group_variances) / total_elements

    # Рассчитываем межгрупповую дисперсию
    between_group_variance = sum(group_sizes[g] * (group_means[g] - overall_mean)**2 for g in group_means) / total_elements

    # Доля межгрупповой дисперсии в общей
    between_group_ratio = between_group_variance / overall_variance

    # Возвращаем результаты анализа
    return {
        'num_intervals': num_intervals,
        'groups': groups,
        'group_means': group_means,
        'group_variances': group_variances,
        'group_sizes': group_sizes,
        'overall_mean': overall_mean,
        'overall_variance': overall_variance,
        'within_group_variance': within_group_variance,
        'between_group_variance': between_group_variance,
        'between_group_ratio': between_group_ratio
    }

# Создаем и анализируем разбиения
results_s_minus_1 = analyze_grouping(s-1, height_data, min_height, max_height)
results_s = analyze_grouping(s, height_data, min_height, max_height)
results_s_plus_1 = analyze_grouping(s+1, height_data, min_height, max_height)

# Функция для вывода результатов
def print_results(results):
    print(f"\n--- Результаты для {results['num_intervals']} интервалов ---")
    print(f"Общая средняя: {results['overall_mean']:.2f} см")
    print(f"Общая дисперсия: {results['overall_variance']:.2f} см²")

    # Выводим информацию о группах
    for group_name in results['group_means']:
        print(f"\n{group_name}")
        print(f"  Количество стран: {results['group_sizes'][group_name]}")
        print(f"  Средний рост: {results['group_means'][group_name]:.2f} см")
        print(f"  Групповая дисперсия: {results['group_variances'][group_name]:.2f} см²")

    print(f"\nВнутригрупповая дисперсия: {results['within_group_variance']:.2f} см²")
    print(f"Межгрупповая дисперсия: {results['between_group_variance']:.2f} см²")
    print(f"Доля межгрупповой дисперсии в общей: {results['between_group_ratio']*100:.2f}%")

    # Проверка правила сложения дисперсий
    sum_of_variances = results['within_group_variance'] + results['between_group_variance']
    print(f"\nПроверка правила сложения дисперсий:")
    print(f"Внутригрупповая + Межгрупповая = {sum_of_variances:.2f} см²")
    print(f"Общая дисперсия = {results['overall_variance']:.2f} см²")
    print(f"Разница: {abs(sum_of_variances - results['overall_variance']):.4f} см²")

# Выводим результаты для всех разбиений
print_results(results_s_minus_1)
print_results(results_s)
print_results(results_s_plus_1)

# Сравниваем разбиения по доле межгрупповой дисперсии
print("\n--- Сравнение разбиений ---")
print(f"Разбиение на {s-1} интервалов: {results_s_minus_1['between_group_ratio']*100:.2f}% межгрупповой дисперсии")
print(f"Разбиение на {s} интервалов: {results_s['between_group_ratio']*100:.2f}% межгрупповой дисперсии")
print(f"Разбиение на {s+1} интервалов: {results_s_plus_1['between_group_ratio']*100:.2f}% межгрупповой дисперсии")

# Определяем лучшее разбиение
best_ratio = max(results_s_minus_1['between_group_ratio'],
                 results_s['between_group_ratio'],
                 results_s_plus_1['between_group_ratio'])

if best_ratio == results_s_minus_1['between_group_ratio']:
    best_intervals = s-1
elif best_ratio == results_s['between_group_ratio']:
    best_intervals = s
else:
    best_intervals = s+1

print(f"\nЛучшее разбиение: {best_intervals} интервалов с долей межгрупповой дисперсии {best_ratio*100:.2f}%")

# Визуализируем результаты
plt.figure(figsize=(12, 8))

ratios = [
    results_s_minus_1['between_group_ratio'] * 100,
    results_s['between_group_ratio'] * 100,
    results_s_plus_1['between_group_ratio'] * 100
]

bars = plt.bar([s-1, s, s+1], ratios, color=['blue', 'green', 'orange'])
plt.xlabel('Количество интервалов')
plt.ylabel('Доля межгрупповой дисперсии в общей, %')
plt.title('Сравнение разбиений по доле межгрупповой дисперсии')
plt.xticks([s-1, s, s+1])

# Добавляем значения на вершине столбцов
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.2f}%',
             ha='center', va='bottom')

# Подсвечиваем лучшее разбиение
best_index = [s-1, s, s+1].index(best_intervals)
bars[best_index].set_color('red')

plt.tight_layout()

# Создаем сравнительную диаграмму дисперсий для трех разбиений
plt.figure(figsize=(14, 8))

intervals = [s-1, s, s+1]
within_group_variances = [
    results_s_minus_1['within_group_variance'],
    results_s['within_group_variance'],
    results_s_plus_1['within_group_variance']
]
between_group_variances = [
    results_s_minus_1['between_group_variance'],
    results_s['between_group_variance'],
    results_s_plus_1['between_group_variance']
]

width = 0.35
x = np.arange(len(intervals))

plt.bar(x, within_group_variances, width, label='Внутригрупповая дисперсия')
plt.bar(x, between_group_variances, width, bottom=within_group_variances, label='Межгрупповая дисперсия')

plt.axhline(y=results_s['overall_variance'], color='r', linestyle='--',
           label=f'Общая дисперсия ({results_s["overall_variance"]:.2f} см²)')

plt.xlabel('Количество интервалов')
plt.ylabel('Дисперсия (см²)')
plt.title('Сравнение внутригрупповой и межгрупповой дисперсий для разных разбиений')
plt.xticks(x, intervals)
plt.legend()

# Добавляем процентное соотношение
for i, (within, between) in enumerate(zip(within_group_variances, between_group_variances)):
    total = within + between
    plt.text(i, within/2, f'{within/total*100:.1f}%', ha='center', va='center')
    plt.text(i, within + between/2, f'{between/total*100:.1f}%', ha='center', va='center')

plt.tight_layout()
plt.show()
```

Проверить соответствие выборки закону распределения, которое предположили в п.2 ( используя критерии Пирсона, Колмогорова, ...)

```python
from scipy.stats import norm, chi2
mean_height = np.mean(height_values)
std_height = np.std(height_values, ddof=1)
min_height = min(height_values)
max_height = max(height_values)
n = len(height_values)

print(f"Описательные статистики:")
print(f"Количество наблюдений: {n}")
print(f"Среднее значение: {mean_height:.2f} см")
print(f"Стандартное отклонение: {std_height:.2f} см")
print(f"Минимальное значение: {min_height:.2f} см")
print(f"Максимальное значение: {max_height:.2f} см")

# 1. Визуальные методы проверки нормальности

# 1.1 Гистограмма распределения с наложенной кривой нормального распределения
plt.figure(figsize=(12, 8))

# Построение гистограммы
sns.histplot(height_values, kde=False, bins=20, stat='density', color='skyblue', alpha=0.7)

# Наложение теоретической кривой нормального распределения
x = np.linspace(min_height, max_height, 1000)
y = norm.pdf(x, mean_height, std_height)
plt.plot(x, y, 'r-', linewidth=2, label='Нормальное распределение')

plt.title('Гистограмма распределения роста с наложенной кривой нормального распределения')
plt.xlabel('Рост (см)')
plt.ylabel('Плотность')
plt.legend()
plt.grid(True, alpha=0.3)

# 1.2 Q-Q Plot (график квантиль-квантиль)
plt.figure(figsize=(10, 8))
stats.probplot(height_values, dist="norm", plot=plt)
plt.title('Q-Q Plot для проверки нормальности распределения')
plt.grid(True, alpha=0.3)

# 2. Статистические тесты

# 2.1 Тест Шапиро-Уилка (лучше работает для небольших выборок)
shapiro_test = stats.shapiro(height_values)
print(f"\nТест Шапиро-Уилка:")
print(f"Статистика W: {shapiro_test.statistic:.4f}")
print(f"p-значение: {shapiro_test.pvalue:.4f}")
if shapiro_test.pvalue < 0.05:
    print("Вывод: Распределение не является нормальным (p < 0.05)")
else:
    print("Вывод: Нет оснований отвергать гипотезу о нормальности распределения (p ≥ 0.05)")

# 2.2 Тест Колмогорова-Смирнова
ks_test = stats.kstest(height_values, 'norm', args=(mean_height, std_height))
print(f"\nТест Колмогорова-Смирнова:")
print(f"Статистика D: {ks_test.statistic:.4f}")
print(f"p-значение: {ks_test.pvalue:.4f}")
if ks_test.pvalue < 0.05:
    print("Вывод: Распределение не является нормальным (p < 0.05)")
else:
    print("Вывод: Нет оснований отвергать гипотезу о нормальности распределения (p ≥ 0.05)")

# 2.3 Тест Д'Агостино-Пирсона (проверяет асимметрию и эксцесс)
dagostino_test = stats.normaltest(height_values)
print(f"\nТест Д'Агостино-Пирсона:")
print(f"Статистика: {dagostino_test.statistic:.4f}")
print(f"p-значение: {dagostino_test.pvalue:.4f}")
if dagostino_test.pvalue < 0.05:
    print("Вывод: Распределение не является нормальным (p < 0.05)")
else:
    print("Вывод: Нет оснований отвергать гипотезу о нормальности распределения (p ≥ 0.05)")

# 2.4 Критерий хи-квадрат Пирсона
# Разбиваем данные на интервалы
num_bins = int(np.ceil(np.sqrt(n)))  # Оптимальное количество интервалов по правилу Стёрджеса
hist, bin_edges = np.histogram(height_values, bins=num_bins, density=False)
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

# Ожидаемые частоты по нормальному распределению
expected = n * np.diff(stats.norm.cdf(bin_edges, loc=mean_height, scale=std_height))

# Отбрасываем интервалы с ожидаемой частотой менее 5
valid_indices = expected >= 5
if not all(valid_indices):
    print(f"\nПредупреждение: {np.sum(~valid_indices)} интервалов имеют ожидаемую частоту менее 5.")
    print("Для корректного теста хи-квадрат объединим некоторые интервалы.")

    # Объединяем интервалы с малой ожидаемой частотой
    hist_new = []
    expected_new = []
    bin_centers_new = []

    i = 0
    while i < len(hist):
        if expected[i] < 5:
            # Ищем ближайший интервал для объединения
            if i > 0 and (i == len(hist) - 1 or expected[i-1] < expected[i+1]):
                # Объединяем с предыдущим интервалом
                hist_new[-1] += hist[i]
                expected_new[-1] += expected[i]
                bin_centers_new[-1] = (bin_centers_new[-1] + bin_centers[i]) / 2
            else:
                # Объединяем со следующим интервалом
                temp_hist = hist[i]
                temp_expected = expected[i]
                temp_center = bin_centers[i]

                i += 1
                temp_hist += hist[i]
                temp_expected += expected[i]
                temp_center = (temp_center + bin_centers[i]) / 2

                hist_new.append(temp_hist)
                expected_new.append(temp_expected)
                bin_centers_new.append(temp_center)
        else:
            hist_new.append(hist[i])
            expected_new.append(expected[i])
            bin_centers_new.append(bin_centers[i])
        i += 1

    hist = np.array(hist_new)
    expected = np.array(expected_new)
    bin_centers = np.array(bin_centers_new)

# Расчет статистики хи-квадрат
chi2_stat = np.sum((hist - expected) ** 2 / expected)
# Степени свободы: (количество интервалов - 1 - количество оцениваемых параметров)
# Для нормального распределения оцениваем 2 параметра: среднее и дисперсию
df = len(hist) - 1 - 2
p_value = 1 - stats.chi2.cdf(chi2_stat, df)

print(f"\nКритерий хи-квадрат Пирсона:")
print(f"Количество интервалов после объединения: {len(hist)}")
print(f"Степени свободы: {df}")
print(f"Статистика хи-квадрат: {chi2_stat:.4f}")
print(f"p-значение: {p_value:.4f}")
if p_value < 0.05:
    print("Вывод: Распределение не соответствует нормальному (p < 0.05)")
else:
    print("Вывод: Нет оснований отвергать гипотезу о нормальности распределения (p ≥ 0.05)")

# Визуализация результатов теста хи-квадрат
plt.figure(figsize=(12, 8))
width = np.diff(bin_edges)[0]
plt.bar(bin_centers, hist, width=width*0.8, alpha=0.6, color='skyblue', label='Наблюдаемые частоты')
plt.plot(bin_centers, expected, 'r-o', linewidth=2, label='Ожидаемые частоты')
plt.title('Сравнение наблюдаемых и ожидаемых частот для теста хи-квадрат')
plt.xlabel('Рост (см)')
plt.ylabel('Частота')
plt.legend()
plt.grid(True, alpha=0.3)

# 3. Дополнительные характеристики распределения
skewness = stats.skew(height_values)
kurtosis = stats.kurtosis(height_values)

print(f"\nДополнительные характеристики распределения:")
print(f"Асимметрия: {skewness:.4f} (0 для нормального распределения)")
if skewness < -0.5:
    print("  Распределение имеет отрицательную асимметрию (длинный хвост слева)")
elif skewness > 0.5:
    print("  Распределение имеет положительную асимметрию (длинный хвост справа)")
else:
    print("  Распределение примерно симметрично")

print(f"Эксцесс: {kurtosis:.4f} (0 для нормального распределения)")
if kurtosis < -0.5:
    print("  Распределение более плоское, чем нормальное (платикуртическое)")
elif kurtosis > 0.5:
    print("  Распределение более островершинное, чем нормальное (лептокуртическое)")
else:
    print("  Эксцесс близок к нормальному")

# 4. Общий вывод о соответствии нормальному распределению
print("\n===============================")
print("ОБЩИЙ ВЫВОД О НОРМАЛЬНОСТИ РАСПРЕДЕЛЕНИЯ:")

tests_passed = 0
total_tests = 4  # Всего тестов: Шапиро-Уилка, Колмогорова-Смирнова, Д'Агостино-Пирсона, хи-квадрат

if shapiro_test.pvalue >= 0.05:
    tests_passed += 1
if ks_test.pvalue >= 0.05:
    tests_passed += 1
if dagostino_test.pvalue >= 0.05:
    tests_passed += 1
if p_value >= 0.05:
    tests_passed += 1

if tests_passed == total_tests:
    print("Все статистические тесты подтверждают нормальность распределения.")
elif tests_passed >= total_tests / 2:
    print(f"{tests_passed} из {total_tests} тестов подтверждают нормальность распределения.")
    print("Распределение можно считать приближенно нормальным.")
else:
    print(f"Только {tests_passed} из {total_tests} тестов подтверждают нормальность распределения.")
    print("Распределение значимо отличается от нормального.")

print("===============================")

plt.tight_layout()
plt.show()
```
