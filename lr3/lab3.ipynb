Преобразование словаря в DataFrame

```python
def prepare_data(data):
    variables = ["y", "x1", "x2", "x3", "x4", "x5"]
    result = {var: [] for var in variables}

    for country in data.values():
        for var in variables:
            result[var].append(country[var])

    return result
```

Функции для ручных вычислений

```python
def mean(x):
    return sum(x) / len(x)

def std_dev(x):
    m = mean(x)
    return math.sqrt(sum((xi - m)**2 for xi in x) / len(x))

def covariance(x, y):
    return sum((xi - mean(x)) * (yi - mean(y)) for xi, yi in zip(x, y)) / len(x)

def pearson_corr(x, y):
    return covariance(x, y) / (std_dev(x) * std_dev(y))
```

Расчет корреляционной матрицы

```python
def calculate_correlation_matrix(data):
    variables = list(data.keys())
    matrix = {}

    for var1 in variables:
        matrix[var1] = {}
        for var2 in variables:
            matrix[var1][var2] = round(pearson_corr(data[var1], data[var2]), 2)

    return matrix
```

Основной анализ

```python
def perform_analysis(data):
    corr_matrix = calculate_correlation_matrix(data)
    df = pd.DataFrame(corr_matrix)

    plt.figure(figsize=(12, 8))
    sns.heatmap(df, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
    plt.title("Исходная корреляционная матрица")
    plt.show()

    threshold = 0.4
    weak_vars = [var for var in df.columns if var != "y" and abs(df["y"][var]) < threshold]
    filtered_data = {k: v for k, v in data.items() if k not in weak_vars}

    print(f"Исключенные переменные: {weak_vars}")

    filtered_matrix = calculate_correlation_matrix(filtered_data)
    df_filtered = pd.DataFrame(filtered_matrix)

    plt.figure(figsize=(8, 6))
    sns.heatmap(df_filtered, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
    plt.title("Фильтрованная корреляционная матрица")
    plt.show()

    print("\nКлючевые выводы:")
    print(f"1. Сильнее всего с населением связаны: {list(filtered_data.keys())[1:]}")
    print("2. Межфакторные корреляции:")
    for var1 in filtered_data:
        if var1 == "y": continue
        for var2 in filtered_data:
            if var2 == "y" or var1 == var2: continue
            if abs(filtered_matrix[var1][var2]) > 0.7:
                print(f"   {var1} и {var2}: {filtered_matrix[var1][var2]} (возможна мультиколлинеарность)")

prepared_data = prepare_data(country_data)
perform_analysis(prepared_data)
```
